```
from functools import reduce
```

```
mul_numbers = lambda x, y : x*y
gr8_200 = lambda x : x > 200
add_nos = lambda x, y : x + y
```

```
l1 = list(range(100,105))
l2 = list(range(5))
```

```
out = reduce(add_nos, filter(gr8_200, map(mul_numbers, l1, l2)))
```

```
out = reduce(lambda x, y : x + y, 
             filter(lambda x : x > 200, 
                    map(lambda x, y : x*y, l1, l2)))
```
