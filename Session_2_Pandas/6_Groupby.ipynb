```python
data = {
    "BoyOrGirl": [
        "Boy",
        "Boy",
        "Boy",
        "Girl",
        "Boy",
        "Girl",
        "Boy",
        "Boy",
        "Girl",
        "Girl",
        "Boy",
        "Boy",
    ],
    "Gender": ["M", "F", "M", "M", "F", "F", "F", "M","F", "F", "F", "M"],
    "TestScore": [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70],
}
```

```python
df = pd.DataFrame(data, columns = ['BoyOrGirl', 'TestScore', "Gender" ])
```

```python
df
```

```python
pd.pivot_table(df, index=["BoyOrGirl"], columns=["Gender"], values="TestScore")
```

```python
pd.pivot_table(df, index=["BoyOrGirl"], values="TestScore",
               aggfunc= [np.mean, np.median, min, max, np.std])
```

```python
pd.pivot_table(df, index=["BoyOrGirl"], values="TestScore",
               aggfunc= ["count"])
```
