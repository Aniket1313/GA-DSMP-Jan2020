```python
# Dataframe using Lists
sales = [(str(date.today()), 150, 200, 50),
         (str(date.today()), 200, 210, 90),
         (str(date.today()), 140, 215, 95)]
labels = [str(date.today()), 'Jan', 'Feb', 'Mar']
```

```python
df = pd.DataFrame(sales, columns=labels)
df
```
